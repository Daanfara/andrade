```python
print(weather.shape)
weather_cleaning=weather.dropna(axis=0)
print(weather_cleaning.shape)
```

```python
print(f'The number of deleted data is: {weather.shape[0]-weather_cleaning.shape[0]}')
```

```python
feature1=['air_pressure_9am']
```

```python
auxl=weather_cleaning.shape
print(type(auxl))
print(auxl)
print(auxl[0])
```

```python
df_ml=weather_cleaning.copy()
df_ml.head(10)
```

```python
y=(df_ml[['relative_humidity_3pm']]>25)*1
```

```python
print(type(y))
```

```python
print(y.head(10))
```

```python
x=df_ml[feature1]
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y, test_size=0.30,random_state=248)
```

```python
from sklearn. linear_model import LogisticRegression
model= LogisticRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
predictions=model.predict(xtest)
```

```python
print(predictions[:10])
```

```python
print(ytest[:10])
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_true=ytest,y_pred=predictions)
```
